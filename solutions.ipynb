```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
# Import the merged_clean_ver1.csv file.
file = pd.read_csv('files_for_activities/csv_files/merged_clean_ver1.csv')
file.head()
```

```python
# Check the data types of all columns in the dataframe. 
file.dtypes
```

```python
# On only the float type columns, use the describe method 
file.describe()
```

```python
# What is the difference between population mean and sample mean. 

# The population mean is the mean of the entire population 
# while the sample mean is the mean of a sample of the population.
```

```python
# Specify the column to use for the box plot
column_to_plot = 'target_d'

# Create the box plot
plt.boxplot(file[column_to_plot], 'red')

# Add a title and labels for the x- and y-axes
plt.title('Box Plot of My Column')
plt.xlabel('My Column')
plt.ylabel('Values')

# Show the plot
plt.show()
```

```python
# Specify the column to use for the box plot
column_to_plot = 'target_d'

# Create the box plot
sns.boxplot(x=file[column_to_plot], palette='Set1')

# Add a title and labels for the x- and y-axes
plt.title('Box Plot of My Column') 
plt.xlabel('My Column')
plt.ylabel('Values')

# Show the plot
plt.show()
```
